## 0. An end-to-end Scikit-Learn workflow

```python
heart_disease = pd.read_csv('data/heart-disease.csv')
heart_disease.head()
```

## 1. Getting The Data Ready

```python
# Create X (features matrix) and Y (labels or target vector)
X = heart_disease.drop('target', axis=1)
y = heart_disease['target']
```

## 2. Choose the right model and hyperparameters

```python
# our problem is clssification
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier()

# We'll use the default parameters
clf.get_params()
```

## 3. Fit the model to the training data

```python
# We need to split our data into training and test sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
X_train.head() # we have 20% of the data in the test set. # 2D array
```

```python
X_test.head() # we have 80% of the data in the training set. # 2D array
```

```python
y_train # 1D array of zeros and ones
```

```python
y_test # 1D array of zeros and ones
```

```python
clf.fit(X_train, y_train) # fit the model(find patterns in training data)
```

```python
y_preds = clf.predict(X_test) # predict the labels of the test set
y_preds
```

## 4. Evaluate the model on the training data and test data

```python
clf.score(X_train, y_train) # score the model on the training set
```

```python
clf.score(X_test, y_test) # score the model on the test sey
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

print(classification_report(y_test, y_preds))
```

```python
confusion_matrix(y_test, y_preds)
```

```python
accuracy_score(y_test, y_preds)
```

## 5. Improve a model

```python
# 5. Improve a model
# Try different amount of n_estimators
np.random.seed(42)
for i in range(10, 100, 10):
    print(f"Trying model with {i} estimators...")
    clf = RandomForestClassifier(n_estimators=i).fit(X_train, y_train)
    print(f"Model accuracy on test set: {clf.score(X_test, y_test) * 100:.2f}%")
    print("")
```

## 6. Save a model and load it

```python
import pickle

pickle.dump(clf, open("random_forst_model_1.pkl", "wb"))
```

```python
loaded_model = pickle.load(open("random_forst_model_1.pkl", "rb"))
loaded_model.score(X_test, y_test)
```
